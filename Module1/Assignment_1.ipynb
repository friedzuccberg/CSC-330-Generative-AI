**Analysis:**

  Based on the results I gathered from testing, the model repeated the phrase “I love it” in Trial A which had a temperature of 0.1. The reasoning behind this outcome is due to the characteristics of a low temperature. A low temperature in an LLM results in the LLM becoming conservative choosing the highest probability word every time. The word “favorite” in the prompt is indicative of a likeness for coffee, this directs the LLM to an embedding space filled with words of similar meanings and feelings such as love. The high probability of the word love paired with the conservative nature of the model resulted in a repetitive phrase as the model’s output. The model used real words however, the coherentness of the model decreased as the temperature increased. As seen in Trial C the response contained words similar/related to coffee such as “caffeine” and “mix-ins” but the grammatical format made no sense. This can be attributed to the change in the model’s probability distribution. As the temperature of the model increases, the chance of a low probability token being selected increases, resulting in incoherent or gibberish-like sentences. If I was building a medical AI to give prescriptions or advice, I would use a conservative temperature such as 0.1 due to its deterministic and logical nature. This temperature would ensure that the medical AI uses facts and logic when formulating its response. If I was building an AI to write a surrealist dream-journal I would use a temperature of 0.8 due to its creative nature. Models that possess a 0.8 temperature add flair and creativity in their responses while maintaining readability making it the best choice for a surrealist dream-journal.
